```python
import numpy as np
import pandas as pd
from fpdf import FPDF
```

```python
abortion_eps = pd.read_csv("data/cleaned_transcripts/abortion_episodes.csv", index_col= 0)
shows = pd.read_csv("data/shows.csv", index_col=0)
```

```python
def makepdf(ep):
    show = shows[shows.index == ep.show_id].title.values[0]
    header = show + " season " + str(ep.season) + ", episode " + str(ep.episode)
    saveas = "data/abortion transcript pdfs/" + header.replace(" ", "_") + ".pdf"
    
    pdf = FPDF(format='letter') #pdf format
    pdf.add_page() #create new page

    pdf.add_font('Lucida Sans Unicode', '', "data/font/l_10646.ttf", uni=True)
    pdf.set_font('Lucida Sans Unicode', size=12)
    
    pdf.multi_cell(200, 10, txt=header, align="C")
    pdf.multi_cell(200, 10, txt="excerpts", align="C")
    pdf.multi_cell(200, 10, txt=str(ep.excerpts), align="L")
    pdf.multi_cell(200, 10, txt="transcript", align="C")
    pdf.multi_cell(200, 10, txt=str(ep.script), align="L")
    pdf.output(saveas)
```

```python
abortion_eps.apply(makepdf, axis=1)
```
